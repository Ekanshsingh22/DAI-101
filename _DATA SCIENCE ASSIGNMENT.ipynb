# CORRELATION MATRIX 

```python
corr_vars = [col for col in data.columns]
corr_matrix = data[corr_vars].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
